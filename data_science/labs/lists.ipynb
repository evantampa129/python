## Lists
### Indexing

A list is a sequenced collection of different objects such as integers, strings, and even other lists as well. The address of each element within a list is called an index. An index is used to access and refer to items within a list.

```python
# Create a list
L = ["The Bodyguard", 7.0, 1992]
L
```

```python
# Print elements on each index
# Positive and negative indexing
print('Positve indexing:', L[0],'\nNegative indexing:', L[-3])

# Positive and negative indexing
print('Positve indexing:', L[1],'\nNegative indexing:', L[-2])
```

### List Content

```python
# Sample List
["The Bodyguard", 7.0, 1992, [1,2], ("A", 1)]
```

### List Operations

```python
L = ["The Bodyguard", 7.0, 1992, "BG", 1]
L
```

```python
# List Slicing
L[3:5]
```

```python
# Use extend to add elements to list
L = ["The Bodyguard", 7.0]
L.extend(['pop', 10])
L
```

```python
# Use append to add elements to list
# Add one element to the list
L = ["The Bodyguard", 7.0]
L.append(['pop', 10])
L
```

```python
# Use extend to add elements to list
L = [ "The Bodyguard", 7.0]
L.extend(['pop', 10])
L
```

```python
# Use append to add elements to list
L.append(['a','b'])
L
```

### Lists are mutable
We can change them after their creation.

```python
# Change the element based on the index
A = ['disco', 10, 1.2]
print('Before change:', A)
A[0] = 'hard rock'
print('After change', A)
```

```python
# Delete element based on the index
print('Before change:', A)
del(A[0])
print('After change:', A)
```

```python
# Split string, default is by space
'hard rock'.split()
```

```python
# Split string by comma
'A,B,C,D'.split(',')
```

### Copy and Clone List

```python
# Copy by reference the list A

A = ["hard rock", 10, 1.2]
B = A
print('A:', A)
print('B:', B)
```

```python
# Copy by reference
print('B[0]:', B[0])
A[0] = "banana"
print('B[0]:', B[0])
```

```python
# Clone by value
B = A[:]
B
```

```python
# Now if you change A, B will not change
print('B[0]:', B[0])
A[0] = "hard rock"
print('B[0]:', B[0])
```
