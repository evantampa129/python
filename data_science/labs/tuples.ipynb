## Tuples

Definition: a collection used to store multiple items in a single variable
It has two main characteristics: **Ordered:** The items have a defined order, and that order will not change.
**Immutable:** Once a tuple is created, you cannot change, add, or remove items.

```python
# Create a tuple
tuple1 = ("disco", 10, 1.2)
print(tuple1)
type(tuple1)
```

### Indexing

Each element of a tuple can be accessed via an index. Each element can be obtained by the name of the tuple followed by a square bracket with the index number.

```python
# Print the variable on each index
print(tuple1[0])
print(tuple1[1])
print(tuple1[2])
```

```python
# Print the type of value on each index
print(type(tuple1[0]))
print(type(tuple1[1]))
print(type(tuple1[2]))
```

```python
# Use negative indexing to get value of the third element
tuple1[-1]
```

```python
# Use negative indexing to get value of the second element
tuple1[-2]
```

```python
# Use negative indexing to get value of the first element
tuple1[-3]
```

## Concatenate Tuples

We can concatenate tuples using the + sign. 

```python
# Concatenate Tuples
tuple2 = tuple1 + ("hard rock", 10)
tuple2
```

## Slicing

We can slice tuples, obtaining new tuples with the corresponding elements.

```python
# Slice from index 0 to 2
tuple2[0:3]
```

```python
# Slice from index 3 to 4
tuple2[3:5]
```

```python
# Get the length of the tuple
len(tuple2)
```

## Sorting

```python
# Sorted tuple
Ratings = (0, 9, 6, 5, 10, 8, 9, 6, 2)
RatingSorted = sorted(Ratings)
RatingSorted
```

## Nested Tuple

```python
# Create a nested tuple
NestedT = (1, 2, ("pop", "rock"), (3, 4), ("disko", (1, 2)))
NestedT
```

```python
# Print element on each index
print("Element 0 of Tuple: ", NestedT[0])
print("Element 1 of Tuple: ", NestedT[1])
print("Element 2 of Tuple: ", NestedT[2])
print("Element 3 of Tuple: ", NestedT[3])
print("Element 4 of Tuple: ", NestedT[4])
```

```python
# Print element on each index, including nest indexes
print("Element 2, 0 of Tuple: ",   NestedT[2][0])
print("Element 2, 1 of Tuple: ",   NestedT[2][1])
print("Element 3, 0 of Tuple: ",   NestedT[3][0])
print("Element 3, 1 of Tuple: ",   NestedT[3][1])
print("Element 4, 0 of Tuple: ",   NestedT[4][0])
print("Element 4, 1 of Tuple: ",   NestedT[4][1])
```

```python
# Access elements in second nested tuple using a third index
NestedT[2][1][0]
```

```python
# Access elements in second nested tuple using a third index
NestedT[2][1][1]
```

```python
# Access elements in fourth nested tuple using a third index
NestedT[4][1][0]
```

```python
# Access elements in fourth nested tuple using a third index
NestedT[4][1][1]
```
