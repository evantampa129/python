## Shopping List

```python
# Create an empty shopping list
shopping_list = ["Watch", "Laptop", "Shoes", "Pen", "Clothes"]
shopping_list.append("Football")
shopping_list
```

```python
# Print first item from the shopping list
print(shopping_list[0])
```

```python
# Print last item from the shopping list
print(shopping_list[-1])
```

```python
# Print the entire shopping list
print(shopping_list)
```

```python
# Print "Laptop" and "shoes"
shopping_list[1:3]
```

```python
# Change item from the shopping list
shopping_list[3] = "Notebook"
shopping_list
```

```python
# Remove 'Clothes' if present
shopping_list.remove("Clothes")
shopping_list
```

```python
## Delete 'Clothes' if present
del(shopping_list[-1])
# del(shopping_list[-1])
shopping_list
```
