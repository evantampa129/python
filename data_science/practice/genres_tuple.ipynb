## Tuples Quiz

```python
genres_tuple = ("pop", "rock", "soul", "hard rock", "soft rock", \
                "R&B", "progressive rock", "disco") 
genres_tuple
```

```python
# FInd the length of the tuple
len(genres_tuple)
```

```python
# Access the element with respect to index 3
genres_tuple[3]
```

```python
# Use slicing to obtain indexes 3,4,5
genres_tuple[3:6]
```

```python
# Find first two elements of the tuple
genres_tuple[0:2]
```

```python
# Find the index of 's' in disco
genres_tuple[-1][2] # find the element with index 2
"disco".find('s') # find the index of the element 's'
```

```python
# Generate a sorted tuple
C_tuple=(-5, 1, -3)
C_sorted = sorted(C_tuple)
C_sorted
```
