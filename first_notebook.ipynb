```python
# (x,y) Plotting with separate table of pairs
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 9, 15]

# data for the table
table_data = [[xi, yi] for xi, yi in zip(x, y)]

# two rows of subplots: top = plot, bottom = table
fig, (ax_plot, ax_table) = plt.subplots(
    2, 1,
    figsize=(6, 6),
    gridspec_kw={'height_ratios': [3, 1]}  # plot taller than table
)

# ---- plot ----
ax_plot.plot(x, y, marker='o')
ax_plot.set_title("My first Plot")
ax_plot.set_xlabel("x")
ax_plot.set_ylabel("y")

# ---- table ----
ax_table.axis("off")  # no axes for the table
tbl = ax_table.table(
    cellText=table_data,
    colLabels=["x", "y"],
    loc="center", 
    cellLoc="center"
)

# make the table a bit taller
tbl.scale(1, 1.5)

plt.tight_layout()
plt.show()
```

```python
# Algorithms: Binary Search

def binary_search(arr, target):
    left, right=0, len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid
        
        elif arr[mid] < target:
            left = mid + 1
        
        else: right = mid + 1

    return -1

nums = [1, 2, 3, 4, 5, 6, 10, 20]
print("Index of 10:", binary_search(nums, 10))
```

```python
# Data Structures: Stack
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)
    
    def pop(self):
        if not self.items:
            return None
        return self.items.pop()

    # make the instance callable to pop
    def __call__(self):
        return self.pop()

stack = Stack()
stack.push(10)
stack.push(20)

print(stack.pop())
stack.push(30)
print(stack())
```

```python
import numpy as np

arr = np.array([1, 3, 2, 5, 4])
cumulative = np.cumsum(arr)

print("Original:", arr)
print("Cumulative:", cumulative)

plt.plot(cumulative)
plt.title("Cumulative Sum")
plt.show()
```
